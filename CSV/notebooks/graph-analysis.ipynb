Load triples và build graph

```python
import json
import networkx as nx

with open("../data/triples.json", "r", encoding="utf-8") as f:
    triples = json.load(f)

G = nx.DiGraph()

for t in triples:
    G.add_edge(
        t["subject"],
        t["object"],
        relation=t["relation"]
    )

print("Nodes:", G.nodes())
print("Edges:", G.edges())
```

Degree Centrality

```python
degree = nx.degree_centrality(G)

print("Degree Centrality:")
for node, score in degree.items():
    print(node, score)
```

Betweenness Centrality

```python
betweenness = nx.betweenness_centrality(G)

print("Betweenness Centrality:")
for node, score in betweenness.items():
    print(node, score)
```

PageRank

```python
pagerank = nx.pagerank(G)

print("PageRank:")
for node, score in pagerank.items():
    print(node, score)
```

Xếp hạng top entities

```python
sorted_pagerank = sorted(
    pagerank.items(),
    key=lambda x: x[1],
    reverse=True
)

print("Top entities:")
for node, score in sorted_pagerank:
    print(node, score)
```

Lưu kết quả phân tích

```python
analysis = {
    "degree": degree,
    "betweenness": betweenness,
    "pagerank": pagerank
}

with open("../data/graph_metrics.json", "w", encoding="utf-8") as f:
    json.dump(analysis, f, ensure_ascii=False, indent=4)
```
